Student Mental Health data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Student Mental Health Survey.csv')
print(df.head())
```

```python
print(df.isnull().sum())
df.dropna(axis=1)
```

```python
print(df.describe())
```

```python
mood_counts = df['How would you describe your general mood this past week'].value_counts().sort_index()
mood_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of General Mood')
plt.xlabel('Mood (1-5)')
plt.ylabel('Number of Students')
plt.show()
```

```python
overwhelmed_counts = df['How often do you feel overwhelmed by daily tasks ( studies, etc)'].value_counts()
overwhelmed_counts.plot(kind='bar', color='lightcoral')
plt.title('Frequency of Feeling Overwhelmed by Daily Tasks')
plt.xlabel('Frequency')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

```python
social_media_counts = df['On average, how much time do you spend daily on social media (Instagram, YouTube, etc.)?'].value_counts()
social_media_counts.plot(kind='bar', color='lightgreen')
plt.title('Time Spent Daily on Social Media')
plt.xlabel('Time Spent')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()

```

```python
cross_tab = pd.crosstab(df['How would you describe your general mood this past week'], df['On average, how much time do you spend daily on social media (Instagram, YouTube, etc.)?'])
cross_tab.plot(kind='bar', stacked=True)
plt.title('General Mood vs. Social Media Usage')
plt.xlabel('General Mood')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.legend(title='Social Media Usage')
plt.show()
```
